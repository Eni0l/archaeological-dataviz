```python
import seaborn as sns
import matplotlib.pyplot as plt    # basically this is used to import plots into python
plt.style.use("seaborn")
```

```python
trench_a = df[df['Method'] == 'Trench A']  # this is used to select only data from trench A
```

```python
columns = [col for col in df.columns if 'Unnamed' not in col]  # this is used to remove all the unamed columns from trech A
trench_a = trench_a[columns]
trench_a.head()
```

```python
elements = columns[4:] # selecting only the elements columns, this prepares the data so that we can create a pivot table for easy visualization of the heatmaps
elements_full = ['Magnesium', 'Boron Aluminide', 'Aluminium', 'Silicon', 'Phosphorus', 'Sulphur', 'Potassium', 'Calcium', 'Manganese', 'Iron', 'Zinc', 'Gold'] # writing out the full names of the elements
elements_dict = dict(zip(elements, elements_full)) # creating a dictionary to pair the elements symbols and their full names
elements_dict
```

```python
trench_a_pivot = trench_a.pivot('Column', 'Layer')['P']
trench_a_pivot    # TO HELP SELECT LAYER AND INDEX THE COLUMN OF TRENCH A. INDEX MEANS COUNTING IN NUMBERS SEQUENCIALLY. SO FOR EXAMPLE HERE WE USED ALUMINIUM
```

```python
for element in elements:
  trench_a_pivot = trench_a.pivot('Column', 'Layer' )[element] #THAT BEIGN SAID, WE REPLACE THE AU WITH ALL ELEMENTS
  plt.figure(figsize=(10, 5)) #THIS IS JUST FOR MAP VISUALIZATION
  ax = sns.heatmap(trench_a_pivot, linewidth = 1 , cmap="Blues" ,annot = True)
  sns.color_palette("light:b" , as_cmap=True)
  plt.title(f"{element} Heatmap Using Seaborn Method")
  plt.show()
```

```python
def elemental_content(element, trench='A'):
  """
  Function to highlight all the areas highest in elemental content of the element specified

  Parameters
  ----------
  element : str
      Symbol of element, e.g. Al.
  """
  trench_a_pivot = trench_a.pivot('Column', 'Layer')[element] #creates a pivot table with column and layer for effective visualization in heat maps
  max_index = trench_a_pivot.idxmax() # this selects all the maximum values for each layer and returns the index (column)
  for index, column in enumerate(max_index):
    layer = max_index.index[index]  # selects the layer
    # finally, we print the results
    print(f"For trench {trench}, {elements_dict[element]} highest value in Layer {layer} is on Column {column}: {trench_a_pivot[layer][column]}")
```

```python
# simply enter any element as input to the function
elemental_content('Al')
```
